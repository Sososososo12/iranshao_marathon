```python
data=pd.DataFrame(pd.read_excel(filepath))
```

```python
# 新增一列
data['cut_sentence']=None
```

```python
data
```

```python
# 将series转成list
comment=data['paragragh'].tolist()
```

```python
comment
```

```python
import striptxt
```

```python
comment_outputs=[]
output_line=''
```

```python
for line in comment:
    output_line=striptxt.seg_sentence(line)
    comment_outputs.append(output_line)
```

```python
comment_outputs
```

```python
# 把分词完的list存入dataframe新增的cut_sentence中
data['cut_sentence']=comment_outputs
```

```python
data
```

```python
# 将cut_sentence列中为空字符串的行索引找出，并将其列为list
a=data[data.cut_sentence==''].index.tolist()
```

```python
a
```

```python
for index_num in a:
    data.drop(data.index[index_num],inplace=True)
```
