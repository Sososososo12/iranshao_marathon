```python
import numpy as np
import scipy.stats
import xlrd
import pandas as pd
```

```python
def JS_divergence(p,q):
    M=(p+q)/2
    return 0.5*scipy.stats.entropy(p, M)+0.5*scipy.stats.entropy(q, M)
```
